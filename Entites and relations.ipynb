### Extract and show entities and relations from an ontology 

```python
import rdflib
# Function to extract and display entities and relations from an ontology
def extract_and_display_ontology(ontology_path):
    # Create an RDF graph
    graph = rdflib.Graph()

    # Load the ontology from the specified file
    graph.parse(ontology_path, format='xml')

    # Iterate through triples to extract entities and relations
    entities = set()
    relations = set()

    for subject, predicate, obj in graph:
        entities.add(subject)
        entities.add(obj)
        relations.add(predicate)

    # Display entities and relations
    print("Entities:")
    for entity in entities:
        print(entity)

    print("\nRelations:")
    for relation in relations:
        print(relation)

# Specify the path to your ontology file (replace with your file path)
ontology_file = 'ontology1.owl'

# Call the function to extract and display the ontology
extract_and_display_ontology(ontology_file)
```
